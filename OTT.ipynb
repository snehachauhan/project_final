Converting the Video Streaming Platform(Y/N) columns into integer.

```python
def trans_con(x):
    if x =='Yes':
        return 1
    if x =='No':
        return 0
```

```python
data['Video Streaming Platform(Y/N)']=data['Video Streaming Platform(Y/N)'].apply(trans_con)
```

```python
data
```

Converting the column(Frequency of VSPA) into integer.

```python
def trans_con(x):
    if x =='Daily':
        return 1
    if x =='Monthly':
        return 2
    if x =='Once or twice in a week':
        return 3
    if x =='None of the Above':
        return 0
```

```python
data['Frequency of VSPA']=data['Frequency of VSPA'].apply(trans_con)
```

```python
data
```

Converting the column(Aage) into integer.

```python
def trans_con(x):
    if x =='less than 18':
        return 1
    if x =='18-30':
        return 2
    if x =='31-50':
        return 3
    if x =='more than 50':
        return 4
```

```python
data['Age']=data['Age'].apply(trans_con)
```

```python
data
```

Converting the column(Overall Rating) into integer.

```python
def trans_conn(x):
    if x =='Excellent':
        return 1
    if x =='Good':
        return 2
    if x =='Fair':
        return 3
    if x =='Improvement Needed':
        return 0
```

```python
data['Overall Rating']=data['Overall Rating'].apply(trans_conn)
```
