```python
def trans_conn(x):
    if x =='Excellent':
        return 1
    if x =='Good':
        return 2
    if x =='Fair':
        return 3
    if x =='Improvement Needed':
        return 0
```

```python
data['Overall Experiance']=data['Overall Experiance'].apply(trans_conn)
```

```python
df = pd.read_csv('Video Streaming Platform Analysis.csv', skiprows = range(20))
years = list(map(str, range(1980,2013)))
```
